```python
# Runway Kill-Switch Analytics - Data Generation
# Synthetic financial data for TechFlow AI startup

import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
import warnings
warnings.filterwarnings('ignore')

print("="*70)
print("TECHFLOW AI - RUNWAY KILL-SWITCH ANALYTICS")
print("Generating 18 Months of Financial Data")
print("="*70)

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)
```

```python
# ============================================
# COMPANY PARAMETERS
# ============================================

COMPANY_NAME = "TechFlow AI"
START_DATE = datetime(2024, 1, 1)
END_DATE = datetime(2025, 6, 30)
INITIAL_CASH = 3_500_000  # $3.5M initial cash
MONTHS = 18

print(f"\nCompany: {COMPANY_NAME}")
print(f"Period: {START_DATE.strftime('%Y-%m-%d')} to {END_DATE.strftime('%Y-%m-%d')}")
print(f"Initial Cash: ${INITIAL_CASH:,}")
print(f"Months Tracked: {MONTHS}")

# ============================================
# INITIATIVE DEFINITIONS
# ============================================

initiatives = {
    'CORE_SAAS': {
        'name': 'Core SaaS Platform',
        'start_month': 0,
        'initial_burn': 85000,
        'burn_trend': 1.02,  # 2% monthly growth
        'revenue_start': 45000,
        'revenue_growth': 1.08,  # 8% monthly growth (strong)
        'team_size': 12,
        'success_prob': 0.85,
        'status': 'Active'
    },
    'MOBILE_APP': {
        'name': 'Mobile App Expansion',
        'start_month': 3,
        'initial_burn': 28000,
        'burn_trend': 1.03,
        'revenue_start': 2000,
        'revenue_growth': 1.12,  # 12% growth (good traction)
        'team_size': 4,
        'success_prob': 0.65,
        'status': 'Active'
    },
    'ENTERPRISE_SALES': {
        'name': 'Enterprise Sales Team',
        'start_month': 6,
        'initial_burn': 42000,
        'burn_trend': 1.01,
        'revenue_start': 8000,
        'revenue_growth': 1.15,  # 15% growth (strong pipeline)
        'team_size': 5,
        'success_prob': 0.75,
        'status': 'Active'
    },
    'CONTENT_MARKETING': {
        'name': 'Content Marketing Program',
        'start_month': 2,
        'initial_burn': 15000,
        'burn_trend': 1.00,  # Flat burn
        'revenue_start': 1500,
        'revenue_growth': 1.05,  # 5% growth (slow)
        'team_size': 2,
        'success_prob': 0.50,
        'status': 'Active'
    },
    'APAC_EXPANSION': {
        'name': 'APAC Expansion',
        'start_month': 8,
        'initial_burn': 35000,
        'burn_trend': 1.04,
        'revenue_start': 0,  # No revenue yet!
        'revenue_growth': 1.20,  # 20% growth IF it works
        'team_size': 3,
        'success_prob': 0.35,  # HIGH RISK
        'status': 'Active'
    },
    'AI_FEATURE': {
        'name': 'AI Feature Beta',
        'start_month': 10,
        'initial_burn': 22000,
        'burn_trend': 1.05,
        'revenue_start': 500,
        'revenue_growth': 1.10,
        'team_size': 3,
        'success_prob': 0.40,  # Experimental
        'status': 'Active'
    }
}

print(f"\nðŸ“Š INITIATIVES DEFINED: {len(initiatives)}")
for code, init in initiatives.items():
    print(f"   â€¢ {init['name']} (Month {init['start_month']}, Team: {init['team_size']})")
```

```python
# ============================================
# GENERATE MONTHLY FINANCIAL DATA
# ============================================

print("\n" + "="*70)
print("GENERATING MONTHLY FINANCIAL DATA")
print("="*70)

# Generate date range
dates = pd.date_range(start=START_DATE, periods=MONTHS, freq='MS')

financial_data = []

for month_idx, date in enumerate(dates):
    
    month_data = {
        'month': date,
        'month_number': month_idx + 1,
        'year': date.year,
        'quarter': f"Q{(date.month-1)//3 + 1}",
    }
    
    # Calculate metrics for each initiative
    total_burn = 0
    total_revenue = 0
    
    for code, init in initiatives.items():
        
        # Check if initiative has started
        if month_idx < init['start_month']:
            month_data[f'{code}_burn'] = 0
            month_data[f'{code}_revenue'] = 0
            month_data[f'{code}_team_size'] = 0
            month_data[f'{code}_status'] = 'Not Started'
            continue
        
        # Calculate months since initiative start
        months_active = month_idx - init['start_month']
        
        # Calculate burn (with trend and some randomness)
        burn = init['initial_burn'] * (init['burn_trend'] ** months_active)
        burn = burn * np.random.uniform(0.92, 1.08)  # +/- 8% variance
        
        # Calculate revenue (with trend and randomness)
        revenue = init['revenue_start'] * (init['revenue_growth'] ** months_active)
        revenue = revenue * np.random.uniform(0.85, 1.15)  # +/- 15% variance
        
        # Add success probability impact (underperforming initiatives)
        if init['success_prob'] < 0.5:
            revenue = revenue * np.random.uniform(0.6, 1.0)  # Drag on revenue
        
        # Store data
        month_data[f'{code}_burn'] = round(burn, 2)
        month_data[f'{code}_revenue'] = round(revenue, 2)
        month_data[f'{code}_team_size'] = init['team_size']
        month_data[f'{code}_status'] = init['status']
        
        total_burn += burn
        total_revenue += revenue
    
    # Add totals
    month_data['total_burn'] = round(total_burn, 2)
    month_data['total_revenue'] = round(total_revenue, 2)
    month_data['net_burn'] = round(total_burn - total_revenue, 2)
    
    financial_data.append(month_data)

df_financial = pd.DataFrame(financial_data)

# Calculate cumulative cash
df_financial['cumulative_burn'] = df_financial['net_burn'].cumsum()
df_financial['cash_balance'] = INITIAL_CASH - df_financial['cumulative_burn']

# Calculate runway remaining (months of cash left)
df_financial['runway_months'] = (
    df_financial['cash_balance'] / df_financial['net_burn']
).round(1)

print(f"âœ… Generated {len(df_financial)} months of financial data")
```

```python
# ============================================
# DISPLAY SUMMARY
# ============================================

print("\n" + "="*70)
print("FINANCIAL SUMMARY")
print("="*70)

# Current state (latest month)
latest = df_financial.iloc[-1]

print(f"\nðŸ“… CURRENT STATE (Month {int(latest['month_number'])}):")
print(f"   Cash Balance:     ${latest['cash_balance']:>12,.0f}")
print(f"   Monthly Burn:     ${latest['net_burn']:>12,.0f}")
print(f"   Runway Remaining: {latest['runway_months']:>12.1f} months")

print(f"\nðŸ’° REVENUE BY INITIATIVE:")
for code, init in initiatives.items():
    if latest[f'{code}_revenue'] > 0:
        print(f"   {init['name']:<30} ${latest[f'{code}_revenue']:>10,.0f}")

print(f"\nðŸ”¥ BURN BY INITIATIVE:")
for code, init in initiatives.items():
    if latest[f'{code}_burn'] > 0:
        print(f"   {init['name']:<30} ${latest[f'{code}_burn']:>10,.0f}")

# Identify initiatives in kill zone
print(f"\nâš ï¸  INITIATIVES IN DANGER ZONE:")
for code, init in initiatives.items():
    revenue = latest[f'{code}_revenue']
    burn = latest[f'{code}_burn']
    
    if burn > 0:
        roi = ((revenue - burn) / burn) * 100 if burn > 0 else 0
        
        if roi < -50:
            print(f"   ðŸ”´ {init['name']}: ROI = {roi:.1f}% (KILL ZONE)")
        elif roi < 0:
            print(f"   ðŸŸ¡ {init['name']}: ROI = {roi:.1f}% (PAUSE ZONE)")
```

```python
# ============================================
# SAVE DATA
# ============================================

print("\n" + "="*70)
print("SAVING DATASETS")
print("="*70)

# Save main financial data
df_financial.to_csv("D:/Projects/End-to-end projects/7. Runway Kill-Switch Analytics/Data/monthly_financials.csv", index=False)
print("âœ… Saved: monthly_financials.csv")

# Create initiative master table
initiative_master = []
for code, init in initiatives.items():
    initiative_master.append({
        'initiative_id': code,
        'initiative_name': init['name'],
        'start_month': init['start_month'],
        'team_size': init['team_size'],
        'success_probability': init['success_prob'],
        'status': init['status'],
        'initial_monthly_burn': init['initial_burn'],
        'initial_monthly_revenue': init['revenue_start']
    })

df_initiatives = pd.DataFrame(initiative_master)
df_initiatives.to_csv("D:/Projects/End-to-end projects/7. Runway Kill-Switch Analytics/Data/initiative_master.csv", index=False)
print("âœ… Saved: initiative_master.csv")

# Create detailed burn/revenue long format
detailed_data = []

for idx, row in df_financial.iterrows():
    for code, init in initiatives.items():
        if row[f'{code}_burn'] > 0 or row[f'{code}_revenue'] > 0:
            detailed_data.append({
                'month': row['month'],
                'month_number': row['month_number'],
                'initiative_id': code,
                'initiative_name': init['name'],
                'monthly_burn': row[f'{code}_burn'],
                'monthly_revenue': row[f'{code}_revenue'],
                'net_burn': row[f'{code}_burn'] - row[f'{code}_revenue'],
                'team_size': row[f'{code}_team_size'],
                'status': row[f'{code}_status']
            })

df_detailed = pd.DataFrame(detailed_data)
df_detailed.to_csv("D:/Projects/End-to-end projects/7. Runway Kill-Switch Analytics/Data/initiative_monthly_detail.csv", index=False)
print("âœ… Saved: initiative_monthly_detail.csv")

print("\n" + "="*70)
print("âœ… DATA GENERATION COMPLETE!")
print("="*70)
print(f"\nNext Steps:")
print(f"1. Load data into PostgreSQL")
print(f"2. Run SQL analysis queries")
print(f"3. Build Excel scenario model")
print(f"4. Create Python ROI uncertainty analysis")
print(f"5. Build Power BI dashboard")
```
